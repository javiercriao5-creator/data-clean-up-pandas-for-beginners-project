#### Ejercicio 01. Análisis de datos básico con estructuras nativas de python.
Crea una lista con las calificaciones de 5 estudiantes: [8.5, 9.2, 7.8, 8.9, 10].

- Calcula el promedio de las calificaciones.

```python
# Código
calificaciones = [8.5, 9.2, 7.8, 8.9, 10]
promedio = sum(calificaciones) / len(calificaciones)
print(f"La lista de calificaciones es: {calificaciones}")
print(f"El promedio de las calificaciones es: {promedio:.2f}")
```

- Encuentra la calificación más alta y la más baja.

```python
# Código
calificacion_maxima = max(calificaciones)
calificacion_minima = min(calificaciones)
print(f"La calificación máxima es: {calificacion_maxima}")
print(f"La calificación mínima es: {calificacion_minima}")
```

#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# TODO
precio_maximo = ds['price'].max()
casa_mas_cara = ds[ds['price'] == precio_maximo].iloc[0]
direccion_casa = casa_mas_cara['address']
precio_casa = casa_mas_cara['price']
print(f"La casa con dirección en {direccion_casa} es la más cara y su precio es de {precio_casa} EUR")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
# TODO
precio_minimo = ds['price'].min()
casa_mas_barata = ds[ds['price'] == precio_minimo].iloc[0]
direccion_casa_barata = casa_mas_barata['address']
precio_casa_barata = casa_mas_barata['price']
print(f"La casa con dirección en {direccion_casa_barata} es la más barata y su precio es de {precio_casa_barata} EUR")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

# Definir la columna de ubicación y la población objetivo
COLUMNA_UBICACION = "level5"
POBLACION_OBJETIVO = "Arroyomolinos (Madrid)"
COLUMNA_PRECIO = "price" # Basado en el Ejercicio 07. Si da error, usar 'Price'.

# 1. SEGMENTAR LOS DATOS por la población
df_arroyomolinos = ds[ds[COLUMNA_UBICACION] == POBLACION_OBJETIVO]

# Verificar si hay datos antes de graficar (por si la población no existe)
if df_arroyomolinos.empty:
    print(f"Error: No se encontraron datos para la población '{POBLACION_OBJETIVO}' en la columna '{COLUMNA_UBICACION}'.")
else:
    precios_arroyomolinos = df_arroyomolinos[COLUMNA_PRECIO]
    media_precio = precios_arroyomolinos.mean()

    # 2. CREACIÓN DEL HISTOGRAMA
    plt.figure(figsize=(10, 6))
    
    # bins=20 crea 20 barras. edgecolor='black' mejora la visibilidad.
    plt.hist(precios_arroyomolinos, bins=20, edgecolor='black', color='lightcoral', alpha=0.8) 

    # 3. CONFIGURACIÓN DEL GRÁFICO
    plt.title(f'Distribución de Precios en {POBLACION_OBJETIVO}')
    plt.xlabel('Precio (EUR)')
    plt.ylabel('Frecuencia (Número de Propiedades)')
    
    # Añadir una línea vertical en la media (vista en Ejercicio 07)
    plt.axvline(media_precio, color='red', linestyle='dashed', linewidth=2, 
                label=f"Media: {media_precio:,.2f} EUR")
    
    plt.legend()
    plt.grid(axis='y', alpha=0.5)

    # 4. IMPRIMIR EL GRÁFICO
    plt.show()
# Trazar el histograma de precios
```
